```python
from PIL import Image
```

```python
img = Image.open('/content/Cute_dog.jpg')
img_resized = img.resize((200,200))
```

```python
img_resized.save('resized_image.jpg')
```

```python
img_res = mpimg.imread('/content/resized_image.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```
