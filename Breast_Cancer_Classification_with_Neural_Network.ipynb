```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
dataframe = pd.DataFrame(dataset.data,columns = dataset.feature_names)
```

```python
dataframe.head()
```

```python
dataframe['label'] = dataset.target
```

```python
dataframe.tail()
```

```python
dataframe.info()
```

```python
dataframe.isnull().sum()
```

```python
dataframe.describe()
```

```python
dataframe['label'].value_counts()
```

```python
dataframe.groupby('label').mean()
```

```python
X = dataframe.drop(columns='label',axis=1)
Y = dataframe['label']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
std_X_train = scaler.fit_transform(X_train)
std_X_test = scaler.transform(X_test)
```

**Building Neural Network**

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape =(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
model.compile (optimizer='adam',
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])
```

```python
history = model.fit(std_X_train,Y_train,validation_split=0.1,epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='upper right')
```

```python
loss,accuracy = model.evaluate(std_X_test,Y_test)
print(accuracy)
```
